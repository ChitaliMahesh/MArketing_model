```python
import  pandas as pd
import numpy  as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression  
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV

from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

from scipy.stats import zscore
```

```python
df=pd.read_csv(r"C:\Users\mahes\OneDrive\Desktop\python\Datasets\bank-full.csv",delimiter=";")
df
```

```python
df.describe()
```

```python
z_score=zscore(df["age"])
threshold=3
index=np.where(np.abs(z_score)> threshold)
index
```

```python
df=pd.get_dummies(df,columns=['job'],drop_first=True)
```

```python
df
```

```python
df['education'].value_counts().to_dict()
```

```python
df['education'].replace({'secondary': 2, 'tertiary': 3, 'primary': 1, 'unknown': 0},inplace=True)
```

```python
df['default'].value_counts().to_dict()
```

```python
df['default'].replace({'no': 0, 'yes': 1},inplace=True)
```

```python
df['housing'].value_counts().to_dict()
```

```python
df['housing'].replace({'yes': 1, 'no': 0},inplace=True)
```

```python
df['loan'].value_counts().to_dict()
```

```python
df['loan'].replace({'no': 0, 'yes': 1},inplace=True)
```

```python
df['contact'].value_counts().to_dict()
```

```python
df['contact'].replace({'cellular': 1, 'unknown': 0, 'telephone': 2},inplace=True)
```

```python
df=pd.get_dummies(df,columns=['month'],drop_first=True)
```

```python
df['marital'].value_counts().to_dict()
```

```python
df['marital'].replace({'married': 1, 'single': 0, 'divorced': 2},inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['poutcome'].value_counts().to_dict()
```

```python
df['poutcome'].replace({'unknown': 1, 'failure': 0, 'other': 2, 'success': 3},inplace=True)
```

```python
df['y'].value_counts().to_dict()
```

```python
df['y'].replace({'no': 0, 'yes': 1},inplace=True)
```

```python
x=df.drop('y',axis=1)
y=df['y']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
Log_model=LogisticRegression()
Log_model.fit(x_train,y_train)
```

```python
# Training 

y_pred_train=Log_model.predict(x_train)

acc_value=accuracy_score(y_train,y_pred_train)
print("Accuracy score",acc_value)

cnf_matrix=confusion_matrix(y_train,y_pred_train)
print("COnfusion matrix\n",cnf_matrix)

classification_value=classification_report(y_train,y_pred_train)
print("Classification report \n",classification_value)
```

```python
# Testing 

y_pred_test=Log_model.predict(x_test)

acc_value=accuracy_score(y_test,y_pred_test)
print("Accuracy score",acc_value)

cnf_matrix=confusion_matrix(y_test,y_pred_test)
print("COnfusion matrix\n",cnf_matrix)

classification_value=classification_report(y_test,y_pred_test)
print("Classification report \n",classification_value)
```

```python
# By using decision tree

dec_model=DecisionTreeClassifier()
dec_model.fit(x_train,y_train)
```

```python
# Training 

y_pred_train=dec_model.predict(x_train)

acc_value=accuracy_score(y_train,y_pred_train)
print("Accuracy score",acc_value)

cnf_matrix=confusion_matrix(y_train,y_pred_train)
print("COnfusion matrix\n",cnf_matrix)

classification_value=classification_report(y_train,y_pred_train)
print("Classification report \n",classification_value)
```

```python
# Testing 

y_pred_test=dec_model.predict(x_test)

acc_value=accuracy_score(y_test,y_pred_test)
print("Accuracy score",acc_value)

cnf_matrix=confusion_matrix(y_test,y_pred_test)
print("COnfusion matrix\n",cnf_matrix)

classification_value=classification_report(y_test,y_pred_test)
print("Classification report \n",classification_value)
```
